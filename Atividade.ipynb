### Atividade Complementar


Imagine que você é o responsável pelo controle de estoque e pretende criar um sistema que registre os pedidos das pessoas ou informe que não há estoque suficiente para atender a demanda.

Nesse caso digamos que os produtos que estão disponíveis no estoque sejam Sabão (com 75 unidades disponíveis); Detergente (com 30 unidades disponíveis) e Desinfetante (com 50 unidades disponíveis).

Crie um código em que seja pedido ao usuário que informe o produto e a quantidade desejada.

Ao final o sistema precisa analisar o pedido com os valores disponíveis no estoque e informar se o pedido foi feito com sucesso ou não foi possível atender o pedido devido à quantidade em estoque ser inferior à demandada.

```python
estoque = {"Sabão": 75, "Detergente": 30, "Desinfetante": 50}

produto_desejado = input("Informe o produto desejado (Sabão, Detergente ou Desinfetante): ")
quantidade_desejada = int(input("Informe a quantidade desejada: "))

# Verificando se o produto existe no estoque
if produto_desejado in estoque:
    quantidade_disponivel = estoque[produto_desejado]

    # Analisando se há estoque suficiente
    if quantidade_desejada <= quantidade_disponivel:
        print("\nPedido feito com sucesso!")
        print(f"Você pediu {quantidade_desejada} unidade(s) de {produto_desejado}.")
    else:
        print("\nNão foi possível atender o pedido.")
        print(f"A quantidade em estoque de {produto_desejado} ({quantidade_disponivel} un.) é inferior à demandada ({quantidade_desejada} un.).")
else:
    print("\nProduto não encontrado. Certifique-se de digitar exatamente como indicado (respeitando maiúsculas e acentos).")
```
